```python
from matplotlib.pyplot import legend, xlabel, ylabel, title, xlim, ylim, plot, show, figure
from pylab import *
from hapi import *

# Download methane lines in 4750-5000 cm-1 wavenumber range
fetch('CH4',6,1,4750,5000)

# Set simulation parameters
Temp = [296., 350., 400.,] # Temperatures, K
Pres = [1., 5., 10.] # Pressures, atm
Reso = 0.1 # Instrument resolution, cm-1
Leng = 300 # Optical path length, cm

# Overlap plots for different temperature and pressure
leg = []
for t,p in zip(Temp,Pres):
    # Calculate absotprion coefficient using Lorentz profile
    nu,coef = absorptionCoefficient_Lorentz(
        SourceTables='CH4',
        HITRAN_units=False,
        Environment={'T':t,'p':p})
    # Calculate absorption spectrum
    nu,coef = absorptionSpectrum(nu,coef,
        Environment={'l':Leng})
    # Convolve spectrum with instrumental function (rectangular by default)
    nu,coef,_,_,_ = convolveSpectrum(
        nu,coef,Resolution=Reso)
    # Plot result of calculation
    plot(nu,coef)
    leg.append('T=%.fK, p=%.fatm'%(t,p))
    
# Set information for plotting
legend(leg,framealpha=0.9)
xlabel('Wavenumber / $cm^{-1}$') 
ylabel('Absorption')
t = 'Modeled absorption spectra for $CH_4$.\n' + 'Resolution: %.1f $cm^{-1}$'%Reso + ', Path length: %.1f $cm$'%Leng
title(t)
xlim(4820,4910)
ylim(0,0.025)

show()

```
