```python
from hapi import *
import matplotlib.pyplot as plt

# Fetch the ozone data from the HITRAN database
fetch('O3', 3, 1, 2000, 2100)

# Calculate absorption coefficients
nu1, coef1 = absorptionCoefficient_Voigt(((3,1),), 'O3', OmegaStep=0.01, HITRAN_units=False, GammaL='gamma_self', Environment={'p':1,'T':296.})
nu2, coef2 = absorptionCoefficient_Voigt(((3,1),), 'O3', OmegaStep=0.01, HITRAN_units=False, GammaL='gamma_self', Environment={'p':5,'T':296.})
nu3, coef3 = absorptionCoefficient_Voigt(((3,1),), 'O3', OmegaStep=0.01, HITRAN_units=False, GammaL='gamma_self', Environment={'p':1,'T':500.})
nu4, coef4 = absorptionCoefficient_Voigt(((3,1),), 'O3', OmegaStep=0.01, HITRAN_units=False, GammaL='gamma_self', Environment={'p':5,'T':500.})

# Create a figure and a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Plot the absorption coefficients
axs[0, 0].plot(nu1, coef1)
axs[0, 0].set_title('O3 k(w): p=1 atm, T=296K')

axs[0, 1].plot(nu2, coef2)
axs[0, 1].set_title('O3 k(w): p=5 atm, T=296K')

axs[1, 0].plot(nu3, coef3)
axs[1, 0].set_title('O3 k(w): p=1 atm, T=500K')

axs[1, 1].plot(nu4, coef4)
axs[1, 1].set_title('O3 k(w): p=5 atm, T=500K')

# Display the figure
plt.tight_layout()
plt.show()
 
```
